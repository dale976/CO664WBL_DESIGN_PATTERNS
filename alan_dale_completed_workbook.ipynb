````python
import matplotlib.pyplot as plt
import numpy as np


# Nicely abstracted structure by which any model can notify any observer (view) of changes in the model
class Subject(object): #Represents what is being 'observed'

    def __init__(self):
        self._observers = [] # This where references to all the observers are being kept
                             # Note that this is a one-to-many relationship: there will be one subject to be observed by multiple _observers

    def attach(self, observer):
        if observer not in self._observers: #If the observer is not already in the observers list
            self._observers.append(observer) # append the observer to the list

    def detach(self, observer): #Simply remove the observer
        try:
            self._observers.remove(observer)
        except ValueError:
            pass

    def notify(self, modifier=None):
        for observer in self._observers: # For all the observers in the list
            if modifier != observer: # Don't notify the observer who is actually doing the updating 
                observer.update(self) # Alert the observers!
                
# Represents the 'data' for which changes will produce notifications to any registered view/observer objects 
class Model(Subject): # Extends the Subject class

    def __init__(self, name):
        Subject.__init__(self)
        self._name = name # Set the name of the model
        self._labels = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
        self._data = [10,8,6,4,2,1]
        self._chart_data = [] # Add a new chart data variable.

    @property #Getter that gets the labels
    def labels(self):
        return self._labels

    @labels.setter #Setter that sets the labels
    def labels(self, labels):
        self._labels = labels
        self.notify() # Notify the observers whenever somebody changes the labels
        
    @property #Getter that gets the data
    def data(self):
        return self._data

    @data.setter #Setter that sets the labels
    def data(self, data):
        self._data = data
        self.notify() # Notify the observers whenever somebody changes the data
    
    @property # Getter for the chart_data, this contains both label and value in one data source
    def chart_data(self):
        return self._chart_data
        
    @chart_data.setter # Setter for chart_data, sets both the label and value
    def chart_data(self, chart_data):
        list_labels = [d['label'] for d in chart_data]
        list_values = [d['value'] for d in chart_data]
        self._labels = list_labels
        self._data = list_values
        self.notify() # Notify the observers
        

# This is the 'standard' view/observer which also acts as an 'abstract' class whereby deriving Bar/Chart/Table views override the update() method
# This 'abstracted' layer is always shown in examples but is important to demonstrate potential polymorphic behaviour of update()
class View():
    
    def __init__(self, name=""):
         self._name = name #Set the name of the Viewer

    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        print("Generalised Viewer '{}' has: \nName = {}; \nLabels = {}; \nData = {}".format(self._name, subject._name, subject._labels, subject._data))

# Table 'chart' viewer/ConcreteObserver - overrides the update() method
class TableView(View):
    
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        fig = plt.figure(dpi=80)
        ax = fig.add_subplot(1,1,1)
        table_data = list(map(list,zip(subject._labels, subject._data)))
        table = ax.table(cellText=table_data, loc='center')
        table.set_fontsize(14)
        table.scale(1,4)
        ax.axis('off')
        plt.show()

# Bar chart viewer/ConcreteObserver - overrides the update() method
class BarView(View):
        
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        objects = subject._labels
        y_pos = np.arange(len(objects))
        performance = subject._data
        plt.bar(y_pos, performance, align='center', alpha=0.5)
        plt.xticks(y_pos, objects)
        plt.ylabel('Usage')
        plt.title('Programming language usage')
        plt.show()
        
# Pie chart viewer/ConcreteObserver - overrides the update() method
class PieView(View):

    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        # Pie chart, where the slices will be ordered and plotted counter-clockwise:
        labels = subject._labels # Use the getter from the model to set the labels
        sizes = subject.data # Use the getter from the model to set the sizes for each pie segmant
        explode = (0.1, 0, 0, 0, 0, 0)  # only "explode" the 1st slice
        fig1, ax1 = plt.subplots()
        ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
        ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show() # Add the "Show" method to render the chart

#Let's create our subjects
m1 = Model("Model 1")
m2 = Model("Model 2") # This is never used!

#Let's create our observers
v1 = View("1: standard text viewer")
v2 = TableView("2: table viewer")
v3 = BarView("3: bar chart viewer")
v4 = PieView("4: pie chart viewer") #Create an instance of the PieView, passing in the name as a prop
####

#Let's attach our observers to the first model
m1.attach(v1)
m1.attach(v2)
m1.attach(v3)
m1.attach(v4) # Attach the pie view to the model, this will attach the observer to the model
####

# Let's just call the notify() method to see all the charts in their unchanged state
m1.notify()

# Now Let's change the properties of our first model
# Change 1 triggers all 4 views and updates their labels
label_list = ['C#','PHP','JavaScript','ASP','Python','Smalltalk'] ### Changed the assignment of the labels here to a variable so it can be re-used
#m1.labels = label_list ### Calling the label setter on the model will notify all observers and update accordingly, as the labels change but the values don't until the next line, the chart will be out of sync. A better way can be done to prevent this!
# Change 2 triggers all 4 views and updates their data
data_list = [1,18,8,60,3,1] ### Changed the assignment of the data here to a variable so it can be re-used
#m1.data = data_list ### Calling the data setter on the model will notify all observers and update accordingly, as the data changes but the labels have now not been updated, the chart will be out of sync. A better way can be done to prevent this!

# Function to convert label data and values into a list of dictionary items
# This data structure will satisfy the new method on the model to update the chart data
# This chart data method will set the labels and values before rendering the list, this will prevent labels and data being out of sync
def create_chart_data_dict_list(label_list, data_list):
    result_list = []

    for label, value in zip(label_list, data_list):
        result_list.append({'label': label, 'value': value})

    return result_list

# Call the new method on the model which will render the charts with both new labels and values
m1.chart_data = create_chart_data_dict_list(label_list, data_list)

# I believe this is a cleaner approach and removes un-wanted re-rendering if both the labels and values need updating at the same time.
# The getter/setter for the individual properties still exist as there will be times when a label/data needs updating in isolation
````

````python
import sys

# Create a simple iterator class
class BasicIteratorWithLength():
    # constructor, which accepts the number of items to return, list
    def __init__(self, numberItems, list):
        self.list = list
        self.numberItems = numberItems
        self.index = 0

        
    def __iter__(self):
        return self
    
    # The next method, will iterate through the list given the index is less than 
    # number defined by the caller
    # stopIteration exception raised if index is greater than the number defined by the caller
    def __next__(self):
        if self.index >= self.numberItems:
            raise StopIteration
        
        item = (self.list[self.index])
        self.index += 1

        return item
            
            

top10books=["Anna Karenina by Leo Tolstoy", "Madame Bovary by Gustave Flaubert", "War and Peace by Leo Tolstoy", 
           "Lolita by Vladimir Nabokov", "The Adventures of Huckleberry Finn by Mark Twain", "Hamlet by William Shakespeare",
           "The Great Gatsby by F. Scott Fitzgerald", "In Search of Lost Time by Marcel Proust", 
           "The Stories of Anton Chekhov by Anton Chekhov", "Middlemarch by George Eliot"]

# Create the new class with a list and a length of 4
myIteratorClass = BasicIteratorWithLength(4, top10books)

# Iterate through the class, printing out the result. Captures the exception
try: 
    for item in myIteratorClass:
        print('Item:', item)
except Exception as e:
    print(e)
    print(sys.exc_info())
````

# The State Design Pattern #

### An object behavioural pattern ##

### Intent ###

"State is a behavioural design pattern that lets an object alter its behaviour when its internal state changes. It appears as if the object changed its class." (Gamma et al., 1995)

A program may be represented by a series of states, with each state behaving differently whislt having no knowledge of the other states that make up the program. 

Characteristics of the state pattern:

- A finite number of states of which a program can be in
- Within a unique state, the program can behave differently
- The program can be switched from one state to another instantaniously
- Strategy can determine if a state can switch to other states
- Switching rules for states (transitions), are finite and predetermined

### Motivation ###

The state pattern manages the behaviour of an object by encapsulating its state into separate classes. The object can change its behaviour when the internal state changes. Each state class is responsible for handling its own conditional logic thus creating better encapsulation within the workspace. 

#### Benefits of this pattern:
- Simplified conditional logic contained within each state class
- Encapsulation naturally occurs as a result of state classes
- Reusability of code and classes
- Maintainability is often easier due to the encapsulation. Making a change in a single state class should have little to no impact on other classes or the context object

#### Consequences of this pattern:
- Increased number of classes due to each state needing its own class
- Complexity in state transitions
- Potential tight coupling between the context object and the state class
- Not every class will utilise the state methods but will require placeholder methods
- Program can scale more easily due to the simplistic nature of adding new state classes

The state design pattern resembles the strategy pattern whereby the behaviour of the program is predetermined by conditional logic. This pattern can be useful in a program where the behaviour needs to reflect a given context. Example real world useages could be a traffic light system, a reservation system, an ordering system or even a video game. 

When a programs behaviour can be determined by predetermined rules, then the state pattern is an abovious choice.

### Implementation
A simple program to print out an element matter. The Matter program has four states: Solid, Liquid, Gas, Plasma. The program will automatically start in a liqiud state.

Simply heating or cooling the element will result in that element changing. Depeding on the current element, it will transition into a different element.

Here the state design pattern can be utilised to manage the element state and allow for changes to be made based on the programs current state. 

An abstract state class is created which will define the abstract methods that exist on the state class and all child classes.
For each element that we define, a new class is made which inherits from the Abstract class. Each element class is responsible for defining the behaviour when the abstract methods are called.

Finally, a context class is created. This is the interface to other areas of the program, such as a GUI

The program can be visualised by the following Class diagram

![alt text](uml_state.png)

````python
# State Abstract State
class State:
    def heat(self):
        pass
    
    def cool(self):
        pass


# Concrete State: SolidState
class SolidState(State):
    def heat(self):
        print("Melting into liquid state")
        return LiquidState()

    def cool(self):
        print("Already in solid state")


# Concrete State: LiquidState
class LiquidState(State):
    def heat(self):
        print("Evaporating into gas state")
        return GasState()

    def cool(self):
        print("Freezing into solid state")
        return SolidState()


# Concrete State: GasState
class GasState(State):
    def heat(self):
        print("Superheating into plasma state")
        return PlasmaState()

    def cool(self):
        print("Condensing into liquid state")
        return LiquidState()

# Concrete State: PlasmaState
class PlasmaState(State):
    def heat(self):
        print("Already in plasma state")

    def cool(self):
        print("Cooling into gas state")
        return PlasmaState()


# Context: Element
class Element:
    def __init__(self):
        self.state = LiquidState()  # Initial state is liquid

    def heat(self):
        self.state = self.state.heat()

    def cool(self):
        self.state = self.state.cool()

# Example usage
element = Element() # Start by creating a Water class as the context
element.cool()  # State Transition: Liquid -> Solid: Output: Freezing into a solid state
element.heat()  # State Transition: Solid -> Liquid: Output: Melting into liquid state
element.heat()  # State Transition: Liquid -> Gas: Output: Evaporating into gas state
element.heat()  # State Transition: Gas -> Plasma: Output: Evaporating into gas state

# Testing the boundaries
# Print the current state
print(element.state) # state is currently PlasmaState Object
# Attempting to transition to an invalid state from the current state
element.heat() # State Transition: Plasma -> Plasma: Output: Already in plasma state
# The object did not change as the state did not get updated. This is determined by logic in encapsulated within the state class


````

# The Façade Design Pattern #

### An object structural pattern ##

### Intent ###

"Provide a unified interface to a set of interfaces in a subsystem. Facade defines a higher-level interface that makes the subsystem easier to use." (Gamma et al., 1995)

The facade design pattern allows for a simple interface to a more complex subsystem with many underlying parts. This pattern is used to hide away the complexities of a system and can be used to separate concerns of different components whilst providing a single interface to interact with.

Characteristics of the Façade pattern:

- A single point of entry for a client
- Simple interface providing functionality from the subsystem
- Deals with interfaces rather than implementation
- Hides internal complexity of a system by providing a simple interface

### Motivation ###

The façade design pattern is used when the underlying complexities of a system are not needed by the client, rather, the façade will provide a simple, single point of entry to interact with. 

#### Benefits of this pattern:
- Simplification by providing a single interface
- Abstraction by hiding away the complexities of the subsystem
- Encapsulation of the interactions between system components
- Decoupling by removing the client from the subsystem 
- Improved maintainability by being able to refactor the subsystem without affecting the interface

#### Consequences of this pattern:
- Complexity of the façade, managing the connectivity of sub components whilst providing a simple, limited scope
- "God Object" façade - The façade class itself can become responsible for far too much and become hard to scale and maintain
- Tight coupling between the façade and the subsystem components. 
- Potential for performance decrease as an extra layer of abstraction is added
- Not very flexible if clients need to access a particular part of the subsystem

The façade pattern closely resembles the adaptor pattern, but rather focuses on interface over implementation. This pattern is best used where a complex system exists but the client requires a single, simple interface. The complexities and interactions of the subsystem are hidden and only limited functionality is provided via the interface to the client.

An example of where this might be used is in a mobile phone application where the underlying platform APIs are required to display information to the user. The platform will provide a way for applications to interface with underlying systems and components where the complexities will be hidden, motion sensors, notifications etc.

Another scenario would be an online streaming music platform. Its graphical user interface(GUI) will provide a way for the user to search for, and play the content. The underlying subsystems will require internet connectivity for accessing and searching through a library of content for the specified content by the user. Additionaly, permission from the platform to use the audio output, this would be an interface provided by the platform.

### Implementation
Below is an example of what a Music Streaming player program might resemble. The Music Player only needs to surface basic actions to the client, the ability to do the following:
    - Play Artist playlist
    - Retrieve albums for an artist
    - Play an album

A façade class is created for the music player providing this limited scope of functionality to the client.
The music player will need to interact with the platform to determine network connectivity to be able to search for music online
The music player will need to interact with the platform to request playing audio

As such, the subsystem will be made up of two classes: PlatformNetworkSystem and PlatformAudioPlayer. Note, these are example classes with basic methods representing what real component classes might do.

These two platform classes are then brought together by another façade, the PlatformFacade which will be responsible for managing the interaction between the system componets. The platform façade will provide simple interface in which these components can be utilised without exposing their internal functionality.


![alt text](music_player_facade.png)

````python
import time
from enum import Enum

# Declare Enum value for player statuses
class PlatformAudioPlayerStatuses(Enum):
    IDLE = 'IDLE'
    PLAYING = 'PLAYING'
    PAUSED = 'PAUSED'
    SEEKING_FORWARD = 'SEEKING_FORWARD'
    SEEKING_REVERSE = 'SEEKING_REVERSE'

# Declare Enum value for player controls
class PlatformPlayerControls(Enum):
    RESUME = 'RESUME'
    PAUSE = 'PAUSE'
    STOP = 'STOP'
    SEEK_FORWARD = 'SEEK_FORWARD'
    SEEK_REVERSE = 'SEEK_REVERSE'

"""
    A facade class providing a simplified interface for interacting with the platform's audio player
    and network system. It encapsulates the initialization of underlying subsystems and provides methods
    for playing audio streams, controlling the audio player, and checking network connectivity.

    Methods:
    - __init__(): Initializes the PlatformFacade by creating instances of PlatformAudioPlayer and
      PlatformNetworkSystem.
    - playAudioStream(audio): Plays an audio stream if a network connection is available; otherwise,
      attempts to reset the connection before playing.
    - controlAudioPlayer(controlRequest): Controls the audio player based on the provided control request,
      including actions such as resume, pause, stop, seek forward, and seek reverse.
    - canPlayAudio(): Checks whether the platform can play audio by verifying network connectivity.
"""
class PlatformFacade:

    def __init__(self):
        print('initialising: Platform Facade')
        self._audioPlayer = PlatformAudioPlayer()
        self._networkSystem = PlatformNetworkSystem()

    def playAudioStream(self, audio):
        if self._networkSystem.has_network_connection():
            self._audioPlayer.play(audio)
        else:
            self._networkSystem.reset_connection()
            self._audioPlayer.play(audio)

    def controlAudioPlayer(self, controlRequest):
        if controlRequest == PlatformPlayerControls.RESUME:
            self._audioPlayer.resume()
        elif controlRequest == PlatformPlayerControls.PAUSE:
            self._audioPlayer.pause()
        elif controlRequest == PlatformPlayerControls.STOP:
            self._audioPlayer.stop()
        elif controlRequest == PlatformPlayerControls.SEEK_FORWARD:
            self._audioPlayer.seek_forwards()
        elif controlRequest == PlatformPlayerControls.SEEK_REVERSE:
            self._audioPlayer.seek_reverse()
        else:
            print('control request not recognised')

    def canPlayAudio(self):
        return self._networkSystem.has_network_connection
        
            
"""
    A class representing the network system of the platform. It provides methods for checking network
    connectivity and resetting the network connection if necessary.

    Methods:
    - __init__(): Initializes the PlatformNetworkSystem with an initial state of having a network connection.
    - has_network_connection(): Checks and returns the current status of network connectivity.
    - reset_connection(): Resets the network connection, simulating the process of establishing a new connection.
"""
class PlatformNetworkSystem:

    def __init__(self):
        self._has_network_connection = True
        print('initialising: platform network system')

    def has_network_connection(self):
        print('Current connectivity is', self._has_network_connection)
        return self._has_network_connection
    
    def reset_connection(self):
        self._has_network_connection = False
        print('attempting to establish a connection...')
        time.sleep(2)
        print('connection established...')
        time.sleep(2)
        print('connection successful')
        self._has_network_connection = True

"""
    A class representing the audio player subsystem of the platform. It provides methods for controlling
    audio playback, such as play, pause, stop, resume, seek forward, and seek reverse, as well as obtaining
    the current status of the audio player.

    Methods:
    - __init__(): Initializes the PlatformAudioPlayer with an initial state of IDLE and no active audio stream.
    - stop(): Stops audio playback, resets the audio stream, and sets the player state to IDLE.
    - play(audio_stream): Starts playing the provided audio stream and sets the player state to PLAYING.
    - pause(): Pauses audio playback and sets the player state to PAUSED.
    - resume(): Resumes audio playback from a paused state and sets the player state to PLAYING.
    - seek_forwards(): Initiates a forward seek operation and sets the player state to SEEKING_FORWARD.
    - seek_reverse(): Initiates a reverse seek operation and sets the player state to SEEKING_REVERSE.
    - get_player_status(): Prints and returns the current state of the audio player.
"""
class PlatformAudioPlayer:
    def __init__(self):
        self._audio_player_state = PlatformAudioPlayerStatuses.IDLE
        self._audio_stream = None
        print('initialising: platform audio system')

    def stop(self):
        self._audio_stream = None
        self._audio_player_state = PlatformAudioPlayerStatuses.IDLE
        print('PlatformAudioPlayer: Stop Playing')

    def play(self, audio_stream):
        self._audio_stream = audio_stream
        self._audio_player_state = PlatformAudioPlayerStatuses.PLAYING
        print('PlatformAudioPlayer: Now Playing ', audio_stream)

    def pause(self):
        self._audio_player_state = PlatformAudioPlayerStatuses.PAUSED
        print('PlatformAudioPlayer: Pause Playing')

    def resume(self):
        self._audio_player_state = PlatformAudioPlayerStatuses.PLAYING
        print('PlatformAudioPlayer: Resume Playing')

    def seek_forwards(self):
        self._audio_player_state = PlatformAudioPlayerStatuses.SEEKING_FORWARD
        print('PlatformAudioPlayer: Seek Forwards')

    def seek_reverse(self):
        self._audio_player_state = PlatformAudioPlayerStatuses.SEEKING_REVERSE
        print('PlatformAudioPlayer: Seek Reverse')

    def get_player_status(self):
        print('audio player is currently', self._audio_player_state)
        return self._audio_player_state
    

"""
    A class representing a music player content library, which stores information about music artists and their albums.

    Attributes:
    - music_artists (dict): A dictionary containing music artists as keys and lists of their albums as values.

    Methods:
    - __init__(): Initializes the MusicPlayerContentLibrary object with pre-defined music artists and albums.
    - getPlaylist(artist): Retrieves the playlist file name for the specified artist.
    - getAlbums(artist): Retrieves the list of album file names for the specified artist.
"""
class MusicPlayerContentLibrary:

    def __init__(self):
        print('initialising: Music Player Content Library')
        self.music_artists = {
        "Taylor Swift": ["Fearless", "1989", "Red"],
        "Beyoncé": ["Lemonade", "B'Day", "4"],
        "Ed Sheeran": ["÷ (Divide)", "x (Multiply)", "+ (Plus)"],
        "Adele": ["21", "25", "19"],
        "Drake": ["Scorpion", "Views", "Take Care"],
        "Lady Gaga": ["The Fame", "Born This Way", "Joanne"],
        "Bruno Mars": ["24K Magic", "Unorthodox Jukebox", "Doo-Wops & Hooligans"],
        "Kanye West": ["My Beautiful Dark Twisted Fantasy", "Yeezus", "The College Dropout"],
        "Rihanna": ["ANTI", "Loud", "Good Girl Gone Bad"],
        "Justin Bieber": ["Purpose", "Believe", "My World 2.0"],
        "Ariana Grande": ["Thank U, Next", "Sweetener", "Dangerous Woman"],
        "Eminem": ["Recovery", "The Eminem Show", "The Marshall Mathers LP"],
        "Coldplay": ["A Head Full of Dreams", "Mylo Xyloto", "X&Y"],
        "Katy Perry": ["Teenage Dream", "PRISM", "Witness"],
        "The Weeknd": ["After Hours", "Starboy", "Beauty Behind the Madness"],
        "Maroon 5": ["V", "Overexposed", "Songs About Jane"],
        "Nicki Minaj": ["Pink Friday", "The Pinkprint", "Queen"],
        "Imagine Dragons": ["Night Visions", "Evolve", "Smoke + Mirrors"],
        "Shawn Mendes": ["Illuminate", "Shawn Mendes", "Handwritten"],
        "Post Malone": ["Hollywood's Bleeding", "Beerbongs & Bentleys", "Stoney"]
    } 

    def getPlaylist(self, artist):
        foundAritst = self.music_artists.get(artist)
        if foundAritst is not None:
            artist_playlist = artist.replace(' ', '') + 'playlist.mp3'
            return artist_playlist
        else:
            print('artist not found')


    def getAlbums(self, artist):
        artistAlbums = self.music_artists.get(artist)
        albumFiles = []
        if artistAlbums is not None:
            print(artistAlbums)
            for album in artistAlbums:
                album = album.replace(' ', '') + '.mp3'
                albumFiles.append(album)
            return albumFiles
        else:
            print('artist not found')

"""
    A facade class providing simplified access to the music streaming functionalities of the platform.
    It encapsulates the initialization of underlying subsystems (PlatformFacade and MusicPlayerContentLibrary)
    and provides methods for playing albums, playing music by artist, and retrieving artist albums.

    Methods:
    - __init__(): Initializes the MusicStreamingFacade by creating instances of PlatformFacade and
      MusicPlayerContentLibrary.
    - playAlbum(albumFile): Plays the specified album file if audio playback is available.
    - playArtistMusic(artistStr): Plays music by the specified artist if audio playback is available,
      using the artist's playlist file retrieved from the content library.
    - getArtistAlbums(artist): Retrieves the list of albums for the specified artist from the content library.
"""        
class MusicStreamingFacade:
    def __init__(self):
        self._platformFacade = PlatformFacade()
        self._contentLibrary = MusicPlayerContentLibrary()

    def playAlbum(self, albumFile):
        if self._platformFacade.canPlayAudio:
            self._platformFacade.playAudioStream(albumFile)
        else:
            print('Unable to play the audio file')

    def playArtistMusic(self, aritstStr):
        if self._platformFacade.canPlayAudio:
            playlistFile = self._contentLibrary.getPlaylist(aritstStr)
            if playlistFile:
                self._platformFacade.playAudioStream(playlistFile)

            else:
                print('Unable to play artist playlist file')
        else:
            print('Unable to play audio. Please check system settings')

    def getArtistAlbums(self, artist):
        albums = self._contentLibrary.getAlbums(artist)
        return albums
            

         
# Simplified usage of the façade pattern
    
# Create an instance of the MusicStreamingFacade, internally this interacts with the platform façade and the music player content library

musicPlayer = MusicStreamingFacade()

# Use the façade instance to get the albums for a given artist, in this case Taylor Swift
taylorSwiftAlbums = musicPlayer.getArtistAlbums('Taylor Swift')

# Print the result
print('\ntaylor swift albums: ', taylorSwiftAlbums)
print('\n')

# Request to play artist music via the façade
musicPlayer.playArtistMusic('Bruno Mars')
print('\n')

# Request to play an album via the façade
musicPlayer.playAlbum(taylorSwiftAlbums[2])


````

## Conclusion
The Façade design pattern is best used when a client needs to interact with a limited part of a system. As shown in the code example above, a complete system can be made up of many complex subsystems which a client does not need to interact with.

The music streaming player is a great example of how a façade can hide away the complexities of an underlying subsystem and bring together the systems and components. From this, a façade provides a single, simple interface for the client.  In the example, the music streaming façade only allows the user the basic controls: search for albums, play a playlist or play an album. The complexities of searching for content, checking system network connectivity and taking over the platforms audio devices are encapsulated within their own classes, hidden away and decoupled from other components. This shows the benefits that the façade pattern can have in a complex system.

It is also worth noting that façades can also interact with other façades and this is a great way of managing larger systems with many subsystem components and services. 

This pattern does come with its disadvantages as the façade file itself can become somewhat of a 'do-it-all' class and become large and hard to maintain very quickly. The façade is also very tightly coupled to the subsystem and its classes, if those change then they still need to provide the same interface to the façade otherwise it will not behave as expected. This can be made even more complicated when façades interact with other façades.
